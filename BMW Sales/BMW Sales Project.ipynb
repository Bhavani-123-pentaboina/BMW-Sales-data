‚ô¶Ô∏è  ANALYSIS

```python
# Sales by Year
sales_by_year = df.groupby('Year')['Sales_Volume'].sum()
print("--- Sales by Year ---")
print(sales_by_year.head())
```

```python
# Top 10 Models
top_models = df.groupby('Model')['Sales_Volume'].sum().sort_values(ascending=False).head(10)
print("--- Top 10 Models ---")
print(top_models)
```

```python
# Sales by Region
sales_by_region = df.groupby('Region')['Sales_Volume'].sum().sort_values(ascending=False)
print("--- Sales by Region ---")
print(sales_by_region)
```

‚ô¶Ô∏è  VISUALIZATIONS

```python
# LINE GRAPH
# Sales Volume Vs Years
plt.plot(sales_by_year.index, sales_by_year.values, marker="o", color="blue")
plt.title("BMW Sales Volume Over Years")
plt.xlabel("Year")
plt.ylabel("Total Sales Volume")
plt.show()
```

```python
# BAR GRAPH
# Total sales Volume vs Model (BAR GRAPH)
top_models.plot(kind="bar", color="skyblue")
plt.title("Top 10 BMW Models by Sales Volume")
plt.xlabel("Model")
plt.ylabel("Total Sales Volume")
plt.show()
```

```python
#  HORIZONTAL BAR GRAPH
sales_by_region = df.groupby("Region")["Sales_Volume"].sum().sort_values()
sales_by_region.plot(kind="barh", color="purple")
plt.title("BMW Sales by Region")
plt.xlabel("Sales Volume")
plt.show()
```

```python
# PIE CHART
sales_by_fuel = df.groupby("Fuel_Type")["Sales_Volume"].sum()
sales_by_fuel.plot(kind="pie", autopct="%1.1f%%", startangle=90, colormap="Set3")
plt.title("BMW Sales Distribution by Fuel Type")
plt.ylabel("")
plt.show()
```

```python
# BOX PLOT
sns.boxplot(data=df, x="Transmission", y="Sales_Volume")
plt.title("Sales Volume Distribution by Transmission Type")
plt.show()
```

```python
# HISTOGRAM
plt.hist(df["Price_USD"], bins=30, color="skyblue", edgecolor="black")
plt.title("Distribution of Car Prices")
plt.xlabel("Price (USD)")
plt.ylabel("Frequency")
plt.show()
```

```python
# HISTOGRAM
plt.hist(df["Mileage_KM"], bins=30, color="lightgreen", edgecolor="black")
plt.title("Distribution of Mileage")
plt.xlabel("Mileage (KM)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Correlation Heatmap
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Heat map
pivot = df.pivot_table(values="Price_USD", index="Fuel_Type", columns="Transmission", aggfunc="mean")
sns.heatmap(pivot, annot=True, cmap="YlGnBu", fmt=".0f")
plt.title("Average Price by Fuel Type & Transmission")
plt.show()
```

```python
# SCATTERPLOT
#Year vs Price
plt.scatter(df['Year'], df['Price_USD'], alpha=0.6, c="brown", edgecolor="k")
plt.xlabel("Year")
plt.ylabel("Price (USD)")
plt.title("Scatter Plot: Year vs Price")
plt.show()
```

```python
# Pie Chart
# Sales by Region
sales_by_region = df.groupby('Region')['Sales_Volume'].sum()
sales_by_region.plot(kind='pie', autopct='%1.1f%%', colormap="Set3")
plt.ylabel("")  # remove y-label
plt.title("BMW Sales Distribution by Region")
plt.show()
```

```python
# Area Chart
# Sales Volume by Year & Region
sales_area = df.groupby(["Year", "Region"])["Sales_Volume"].sum().unstack()
sales_area.plot(kind="area", stacked=True, alpha=0.6, colormap="tab20")
plt.title("BMW Sales Volume Over Years (by Region)")
plt.xlabel("Year")
plt.ylabel("Sales Volume")
plt.legend(title="Region", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```

```python
# Stacked Bar Chart 
# Sales by Fuel & Transmission
stacked_data = df.groupby(["Fuel_Type", "Transmission"])["Sales_Volume"].sum().unstack()
stacked_data.plot(kind="bar", stacked=True, figsize=(10, 6), colormap="Set2")
plt.title("BMW Sales Volume by Fuel Type & Transmission")
plt.xlabel("Fuel Type")
plt.ylabel("Sales Volume")
plt.legend(title="Transmission", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```
